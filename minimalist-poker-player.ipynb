Imports die we gaan gebruiken

```python
import json
from random import random
from websocket import create_connection
```

```python
# maak verbinding met de server
ws = create_connection("ws://ws.sebastiaanvermeulen.nl/pokerroom")

# vraag om aan een tafel te mogen zitten
ws.send(json.dumps({"Table":{"n_players":1,"small_blind":1,"big_blind":2,"stack":40,"game_type":"NoLimit"}}))

print("--------- going to play until > 1000 messages --------")
for _ in range(1000):
    
    # ontvang een JSON-bericht van de server en maak er een Python dictionary van
    message = json.loads(ws.recv())
    print(message)
    
    # Indien de server een reactie verwacht:
    if 'RequestAction' in message:        
        # random response
        rand = random() # uniform [0,1)
        if rand < .5:
            response = json.dumps({'Action':'Call'})
        elif rand < .7:
            response = json.dumps({'Action':{'Raise':2}})
        else:
            response = json.dumps({'Action':'Fold'})
        print(response)
        
        # stuur de reactie naar de server
        ws.send(response)
        
print("--------- hanging up --------")
ws.close()
```
